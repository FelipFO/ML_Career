## Theoretical and Empirical Probability

```python
# Simulate rolling a die
die_faces = [1,2,3,4,5,6]

# Let's create a random generator. The `seed` value 
# is a number used to initialize a pseudorandom number
# generator. It will allow us to get reproducible results
# if we want to run the experiment again.
rng = np.random.RandomState(seed=1)
```

```python
# Randomly choice one of the die faces
choice = rng.choice(die_faces, size=10)
choice
```

```python
# Probability distribution for a die
sns.histplot([1,2,3,4,5,6], bins=list(range(1,8)), stat='probability').set_title(f'Probability Distribution')
plt.show()
```

```python
# Empirical distribution from 100 samples
rng = np.random.RandomState(seed=1)
num_samples = 100
x = rng.choice([1,2,3,4,5,6], num_samples)
sns.histplot(x, bins=list(range(1,8)), stat='probability').set_title(f'Distribution from {num_samples} samples')
plt.show()
```

```python
# Empirical distribution from 100000 samples
rng = np.random.RandomState(seed=1)
num_samples = 10000000
x = rng.choice([1,2,3,4,5,6], num_samples)
sns.histplot(x, bins=list(range(1,8)), stat='probability').set_title(f'Distribution from {num_samples} samples')
plt.show()
```

## Defining our own probability function

Let's code a function which, given a string it may apply different text preprocessing techiniques at randomnly with different probabilities. Basically we want, for a given input name as a string:

- Convert name to lowercase with probability of **0.5**
- Convert name to uppercase with probability of **0.3**
- Leave the name as it is with probability of **0.2**

```python
def name_converter(name, random_generator):
  random_number = random_generator.random()
  if random_number < .5:
    new_name = name.lower()
  elif random_number < .8:
    new_name = name.upper()
  else:
    new_name = name

  return new_name


# We will generate N random names
names_generated = [
  name_converter("Maria", rng)
  for _ in range(1000)
]

# Let's see the empirical distribution
Counter(names_generated)
```

## Working with Sets

```python
# We will use Python sets data structure
X = {1, 2, 3, 4, 5, 6}
Y = {2, 4, 6, 8, 10, 12}
```

### Sets intersection

```python
X.intersection(Y)
```

```python
# Sets intersection is commutative
# X n Y == Y n X
X.intersection(Y) == Y.intersection(X)
```

```python
# Intersection is associative
# (X n Y) n Z == X n (Y n Z)
Z = {1, 2, 5, 100, 1000}
(X.intersection(Y)).intersection(Z) == X.intersection(Y.intersection(Z))
```

### Sets union

```python
X.union(Y)
```

```python
# Sets intersection is commutative
# X u Y == Y u X
X.union(Y) == Y.union(X)
```

```python
# Intersection is associative
# (X u Y) u Z == X u (Y u Z)
Z = {1, 2, 5, 100, 1000}
(X.union(Y)).union(Z) == X.union(Y.union(Z))
```

### Sets Difference

```python
X - Y
```

```python
Y - X
```

```python
# Is Sets difference commutative?
# X - Y == Y - X
X - Y == Y - X
```

```python
# Is Sets difference associative?
# (X - Y) - Z == X - (Y - Z)
(X - Y) - Z == X - (Y - Z)
```
