## CODE GENERATION EXCERCISE #2

###*PROVIDED DATA:*
QUERY:
Give me the code to calculate if a number is prime in Python

TOPIC: Calculation

INPUT DATASET: N/A

### *REQUIRED INFORMATION:*
CODE RESPONSE: Write the function that will complete your given query

```python
import math

def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, math.isqrt(number) + 1):
        if number % i == 0:
            return False
    return True

# Example usage
num = 13

if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

##PLOT GENERATION EXCERCISE #1

###*PROVIDED DATA:*
QUERY:
Plot the tangent wave in python

TOPIC: Plot

INPUT DATASET: data

### *REQUIRED INFORMATION:*
CODE RESPONSE: Write the function that will complete your given query

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from -2*pi to 2*pi
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Calculate tangent values for each x
y = np.tan(x)

# Plot the tangent wave
plt.plot(x, y, label='tan(x)')
plt.title('Tangent Wave')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

PLOT OUTPUT: <font color='red'> The given code prints the tangent function according to the query</font>

COLUMN NAMES: <font color='red'> It doesn't require any dataset, so we don't have any columns </font>
