###*PROVIDED DATA:*
QUERY:
Given the following data, plot a pie chart with the population per country of the 5 most populous considering the total population is 7 billion people, showing just the name of each of them and the percentage. The numbers and the names should be perfectly visible and readable

TOPIC: Plot

INPUT DATASET: data

### *REQUIRED INFORMATION:*
CODE RESPONSE: Write the function that will complete your given query

````python
import matplotlib.pyplot as plt

# Data
gdp_data = {
    'Country': ['United States', 'China', 'Japan', 'Germany', 'India', 'United Kingdom', 'France', 'Brazil', 'Italy', 'Canada', 'South Korea', 'Russia', 'Australia', 'Spain', 'Mexico', 'Indonesia', 'Netherlands', 'Saudi Arabia', 'Turkey', 'Switzerland'],
    'GDP (trillion USD)': [22.675, 17.681, 5.116, 4.418, 3.202, 3.123, 2.714, 2.476, 2.344, 1.986, 1.948, 1.709, 1.488, 1.314, 1.258, 1.119, 0.985, 0.815, 0.794, 0.703],
    'Population (millions)': [331, 1441, 125.5, 83.2, 1393, 68.2, 65.4, 213.9, 60.4, 38.2, 51.8, 145.9, 25.6, 47.3, 128.9, 273.5, 17.5, 35.4, 84.3, 8.7],
    'Continent': ['North America', 'Asia', 'Asia', 'Europe', 'Asia', 'Europe', 'Europe', 'South America', 'Europe', 'North America', 'Asia', 'Europe', 'Oceania', 'Europe', 'North America', 'Asia', 'Europe', 'Asia', 'Asia', 'Europe']
}

# Sort by population in descending order
sorted_data = sorted(zip(gdp_data['Country'], gdp_data['Population (millions)']), key=lambda x: x[1], reverse=True)

# Select the top 5 countries
top_countries = sorted_data[:5]

# Calculate the total population
total_population = sum(country[1] for country in top_countries)

# Calculate percentages and labels
percentages = [(country[0], (country[1] / total_population) * 100) for country in top_countries]

# Plotting
labels, sizes = zip(*percentages)
colors = plt.cm.Paired(range(len(labels)))

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
ax.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.

plt.title('Population Distribution of Top 5 Countries')
plt.show()
````
